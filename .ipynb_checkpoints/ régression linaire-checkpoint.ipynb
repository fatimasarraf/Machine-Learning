Partie 1: Importation des bibliothèques

```python
import numpy as np
import matplotlib.pyplot as plt
```

Partie 2: Génération des données

```python
m = 100
x = np.linspace(0, 10, m).reshape((m, 1))
y = x + np.random.randn(m, 1)
```

Partie 3: Définition des fonctions pour la régression linéaire

```python
def h(theta, x):
    return theta[0] + theta[1] * x

def J(theta, x, y):
    m = len(y)
    return (1/(2*m)) * np.sum((h(theta, x) - y)**2)

def gradient(theta, x, y):
    m = len(y)
    return (1/m) * np.dot(x.T, (h(theta, x) - y))
```

Partie 4: Descente de gradient

```python
def descente_gradient(theta, x, y, alpha, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        theta = theta - alpha * gradient(theta, x, y)
        cost_history[i] = J(theta, x, y)
    return theta, cost_history
```

Partie 5: Entraînement du modèle

```python
theta_initial = np.zeros((2, 1))
alpha = 0.01
iterations = 1000
theta_optimal, cost_history = descente_gradient(theta_initial, x, y, alpha, iterations)
```

Partie 6: Visualisation des données et du modèle

```python
plt.scatter(x, y, color='blue', label='Données')
plt.plot(x, h(theta_optimal, x), color='red', label='Régression linéaire')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Régression linéaire à une seule variable')
plt.legend()
plt.show()
```

Partie 7: Calcul du coefficient de détermination 

```python
y_pred = h(theta_optimal, x)
moy = np.mean(y)
total = np.sum((y - moy)**2)
res = np.sum((y - y_pred)**2)
r2 = 1 - (res / total)
print("Coefficient de détermination R² :", r2)
```

Partie 8: Visualisation de l'évolution du coût

```python
plt.plot(range(iterations), cost_history, color='green')
plt.xlabel('Iterations')
plt.ylabel('Coût J')
plt.title('Évolution du coût lors de la descente de gradient')
plt.show()
```

Partie 3: Définition des fonctions pour la régression linéaire

```python

def h(theta, x1, x2):
    return theta[0] + theta[1] * x1 + theta[2] * x2

# Fonction de coût
def J(theta, x1, x2, y):
    m = len(y)
    return (1/(2*m)) * np.sum((h(theta, x1, x2) - y)**2)

# Gradient
def gradient(theta, x1, x2, y):
    m = len(y)
    grad = np.zeros(theta.shape)
    grad[0] = (1/m) * np.sum(h(theta, x1, x2) - y)
    grad[1] = (1/m) * np.sum((h(theta, x1, x2) - y) * x1)
    grad[2] = (1/m) * np.sum((h(theta, x1, x2) - y) * x2)
    return grad
```

Partie 3: Définition des fonctions pour la régression linéaire

```python
# Fonction de modèle
def h(theta, x1, x2):
    return theta[0] + theta[1] * x1 + theta[2] * x2

# Fonction de coût
def J(theta, x1, x2, y):
    m = len(y)
    return (1/(2*m)) * np.sum((h(theta, x1, x2) - y)**2)

# Gradient
def gradient(theta, x1, x2, y):
    m = len(y)
    grad = np.zeros(theta.shape)
    grad[0] = (1/m) * np.sum(h(theta, x1, x2) - y)
    grad[1] = (1/m) * np.sum((h(theta, x1, x2) - y) * x1)
    grad[2] = (1/m) * np.sum((h(theta, x1, x2) - y) * x2)
    return grad
```

Partie 4: Descente de gradient

```python
def descente_gradient(theta, x1, x2, y, alpha, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        theta = theta - alpha * gradient(theta, x1, x2, y)
        cost_history[i] = J(theta, x1, x2, y)
    return theta, cost_history
```

```python
theta_initial = np.zeros((3, 1))
alpha = 0.01
iterations = 1000
theta_optimal, cost_history = descente_gradient(theta_initial, x1, x2, y, alpha, iterations)
```

Partie 1: Importation des bibliothèques


```python
import numpy as np
import matplotlib.pyplot as plt
```

Partie 2: Génération des données d'apprentissage


```python

m = 100  # Nombre d'exemples
x1 = np.linspace(0, 10, m).reshape((m, 1))  # Première variable
x2 = np.random.randn(m, 1)  # Deuxième variable aléatoire
y = (x1 + np.random.randn(m, 1))**2  # Transformation non-linéaire
```

Partie 3: Définition des fonctions pour la régression polynomiale d'ordre 2


```python
# Fonction de modèle
def h(theta, x1, x2):
    return theta[0] + theta[1] * x1 + theta[2] * x2**2

# Fonction de coût
def J(theta, x1, x2, y):
    m = len(y)
    return (1/(2*m)) * np.sum((h(theta, x1, x2) - y)**2)

# Gradient
def gradient(theta, x1, x2, y):
    m = len(y)
    grad = np.zeros(theta.shape)
    grad[0] = (1/m) * np.sum(h(theta, x1, x2) - y)
    grad[1] = (1/m) * np.sum((h(theta, x1, x2) - y) * x1)
    grad[2] = (1/m) * np.sum((h(theta, x1, x2) - y) * x2**2)  # Gradient par rapport à x2^2
    return grad
```

Partie 4: Descente de gradient


```python
def descente_gradient(theta, x1, x2, y, alpha, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        theta = theta - alpha * gradient(theta, x1, x2, y)
        cost_history[i] = J(theta, x1, x2, y)
    return theta, cost_history
```

Partie 5: Entraînement du modèle


```python
# Initialisation des paramètres
theta_initial = np.zeros((3, 1))

# Paramètres de la descente de gradient
alpha = 0.01
iterations = 1000

# Entraînement du modèle
theta_optimal, cost_history = descente_gradient(theta_initial, x1, x2, y, alpha, iterations)
```

Partie 6: Visualisation des données et du modèle


```python
plt.scatter(x2, x1, y, c='r', label='Données réelles')
plt.scatter(x2, x1, h(theta_optimal, x1, x2), c='g', label='Prédictions du modèle')
plt.xlabel('Feature 2')
plt.ylabel('Feature 1')
plt.title('Régression non-linéaire polynomiale d\'ordre 2')
plt.legend()
plt.show()
```

Partie 7: Calcul du coefficient de détermination

```python
moy = np.mean(y)
ss_total = np.sum((y - moy)**2)
ss_res = np.sum((y - h(theta_optimal, x1, x2))**2)
r2 = 1 - (ss_res / ss_total)
print("Coefficient de détermination R² :", r2)
```

Partie 8: Visualisation de l'évolution du coût


```python
plt.plot(range(iterations), cost_history, color='green')
plt.xlabel('Iterations')
plt.ylabel('Coût J')
plt.title('Évolution du coût lors de la descente de gradient')
plt.show()
```

Partie 1: Importation des bibliothèques


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

Partie 2: Lecture et visualisation des données


```python
# Lecture des données
salary = pd.read_csv('Salary_Data (1).csv')

# Affichage des informations sur les données
print(salary.head())
print(salary.info())
print(salary.describe())
print(salary.columns)
```

Partie 3: Visualisation des données


```python
# Visualisation des données
sns.pairplot(salary)
plt.show()

# Distribution de la variable cible 'Salary'
sns.distplot(salary['Salary'])
plt.show()

# Matrice de corrélation
sns.heatmap(salary.corr(), annot=True, cmap='coolwarm')
plt.show()
```

Partie 4: Prétraitement des données

```python
# Prétraitement des données
X = salary['YearsExperience'].values.reshape(-1, 1)
y = salary['Salary'].values.reshape(-1, 1)

# Ajout d'une colonne de 1 pour l'interception (optionnel pour sklearn mais utile pour comprendre la structure des données)
X = np.hstack((X, np.ones(X.shape)))
```

Partie 5: Division des données en ensembles d'entraînement et de test


```python
# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

Partie 6: Entraînement du modèle


```python
# Entraînement du modèle
lm = LinearRegression()
lm.fit(X_train, y_train)

# Affichage de l'interception
print("Intercept:", lm.intercept_)
```

Partie 7: Prédictions et évaluation du modèle


```python
# Prédictions et évaluation du modèle
predictions = lm.predict(X_test)

# Visualisation des résultats
plt.scatter(y_test, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.show()

# Affichage des mesures d'erreur
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

Partie 1: Importation des bibliothèques


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

Partie 2: Lecture des données et affichage des informations


```python
# Lecture des données
housing_data = pd.read_csv('USA_Housing.csv')

# Affichage des premières lignes et des informations sur les données
print(housing_data.head())
print(housing_data.info())
```

Partie 3: Préparation des données


```python
# Séparation des variables indépendantes (X) et de la variable dépendante (y)
X = housing_data[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 
                  'Avg. Area Number of Bedrooms', 'Area Population']]
y = housing_data['Price']
```

Partie 4: Division des données en ensembles d'entraînement et de test


```python
# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

Partie 5: Entraînement du modèle


```python
# Entraînement du modèle de régression linéaire
lm = LinearRegression()
lm.fit(X_train, y_train)

# Affichage des coefficients
coeff_df = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
print(coeff_df)
```

Partie 6: Prédictions et évaluation du modèle


```python

# Prédictions et évaluation du modèle
predictions = lm.predict(X_test)

# Visualisation des résultats
plt.scatter(y_test, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.show()

# Calcul et affichage des mesures d'erreur
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
