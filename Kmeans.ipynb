```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Création de données

```python
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

Visualisation de données

```python
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow')
plt.title("Visualisation des données")
plt.show()
```

Création de clusters avec K-means


```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

Affichage des résultats


```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))

ax1.set_title('K-Means')
ax1.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow')

plt.show()
```

Étape 1 : Utiliser k-means pour découvrir les clusters dans le jeu de données Iris

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd

# Charger le dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Convertir les données en DataFrame pour une manipulation plus facile
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = y

# Appliquer k-means avec 3 clusters (correspondant aux 3 espèces d'iris)
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(X)

# Ajouter les labels prédits au DataFrame
df['kmeans_labels'] = kmeans.labels_

# Afficher les centres des clusters
print("Centres des clusters :")
print(kmeans.cluster_centers_)

# Visualiser les résultats avec les clusters trouvés par k-means
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=df['species'], cmap='viridis')
plt.title("Véritables labels")

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=df['kmeans_labels'], cmap='viridis')
plt.title("Labels k-means")

plt.show()
```

Étape 2 : Générer des données synthétiques avec des classes qui se chevauchent


```python
from sklearn.datasets import make_blobs

# Générer des données avec des clusters qui se chevauchent
X_blob, y_blob = make_blobs(n_samples=300, centers=4, cluster_std=1.5, random_state=42)

# Appliquer k-means sur ces nouvelles données
kmeans_blob = KMeans(n_clusters=4, n_init=10)
kmeans_blob.fit(X_blob)
labels_blob = kmeans_blob.labels_

# Visualiser les données chevauchantes et les clusters trouvés par k-means
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_blob[:, 0], X_blob[:, 1], c=y_blob, cmap='viridis')
plt.title("Véritables labels (chevauchants)")

plt.subplot(1, 2, 2)
plt.scatter(X_blob[:, 0], X_blob[:, 1], c=labels_blob, cmap='viridis')
plt.title("Labels k-means (chevauchants)")

plt.show()
```

```python
Étape 3 : Explorer les limitations de k-means lorsque les classes se chevauchent
```

```python
# Calculer et afficher les centres des clusters pour les données chevauchantes
print("Centres des clusters pour les données chevauchantes :")
print(kmeans_blob.cluster_centers_)
X_blob_diff, y_blob_diff = make_blobs(n_samples=[100, 300, 200], centers=[[0, 0], [5, 5], [10, 1]], cluster_std=[0.5, 1.5, 0.3], random_state=42)

# Appliquer k-means sur ces nouvelles données
kmeans_blob_diff = KMeans(n_clusters=3, n_init=10)
kmeans_blob_diff.fit(X_blob_diff)
labels_blob_diff = kmeans_blob_diff.labels_

# Visualiser les données et les clusters trouvés par k-means
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_blob_diff[:, 0], X_blob_diff[:, 1], c=y_blob_diff, cmap='viridis')
plt.title("Véritables labels (différentes tailles et formes)")

plt.subplot(1, 2, 2)
plt.scatter(X_blob_diff[:, 0], X_blob_diff[:, 1], c=labels_blob_diff, cmap='viridis')
plt.title("Labels k-means (différentes tailles et formes)")

plt.show()
```
